## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

## 6. Create an interactive line chart showing quantity sold by day for the United Kingdom. Add drop-down boxes for Year and Month that allow you to filter the date range that appears in the chart.

```python
#line chart de time series.  X = día y Y = Quantity sold en UK

#Creamos nuevas columnas con la descomposición de la fecha, y así podemos sacar los días
#Creamos también una columna "UK" para que nos filtre las filas con valor "United Kingdom" en la columna país

data['Year'] = pd.DatetimeIndex(data['InvoiceDate']).year
data['Month'] = pd.DatetimeIndex(data['InvoiceDate']).month
data['Day'] = pd.DatetimeIndex(data['InvoiceDate']).day
uk = data[data['Country']=='United Kingdom']
```

```python
#Creamos el drop down interactivo dejando componer la fecha

@interact(StartMonth = list(uk['Month'].unique()),
          StartYear = list(uk['Year'].unique()),
          EndMonth = list(uk['Month'].unique()),
          EndYear= list(uk['Year'].unique())
         )

def line(StartMonth, StartYear, EndMonth, EndYear):
    data = uk[['Day', 'Month', 'Year', 'Quantity', 'InvoiceDate']].loc[((uk.Year >= StartYear) & 
                                                                        (uk.Year <= EndYear)) & 
                                                                       ((uk.Month >= StartMonth) & 
                                                                        (uk.Month <= EndMonth))
                                                                      ].groupby(['Year',
                                                                                 'Month',
                                                                                 'Day'], as_index=False
                                                                               ).agg({'Quantity':'sum', 
                                                                                      'InvoiceDate':'first'}
                                                                                    )

    data.iplot(kind='line', x='InvoiceDate', y='Quantity', 
               xTitle='Date', yTitle='Amount Sold',
               title='Amount Sold Timeline')
```

## 7. Create an interactive scatter plot that plots number of invoices (x-axis) vs. number of customers (y-axis) and the plot points represent individual products. Add two sliders that control the x and y axis ranges.

```python
#

agg_func = {'InvoiceNo':'nunique',
            'Quantity':'sum',
            'UnitPrice':'mean',
            'Revenue':'sum',
            'CustomerID':'nunique'}

products = uk.groupby('Description').agg(agg_func)

products
```

```python
@interact(Invoices = (products.InvoiceNo.unique().min(), products.InvoiceNo.unique().max()),
          Customers = (products.CustomerID.unique().min(), products.CustomerID.unique().max())
         )

def scatter(Invoices=1, Customers=1):
    data = products.loc[(products.InvoiceNo > Invoices) & (products.CustomerID > Customers)]
    return data.head()
```

```python
@interact(Invoices = (products.InvoiceNo.unique().min(), products.InvoiceNo.unique().max()),
          Customers = (products.CustomerID.unique().min(), products.CustomerID.unique().max())
         )

def scatter(Invoices=1, Customers=1):
    data = products.loc[(products.InvoiceNo > Invoices) & (products.CustomerID > Customers)].head(100)

    data.iplot(kind='scatter', x='InvoiceNo', y='CustomerID', 
               xTitle='Tenure', yTitle='Monthly Charges',
               title='Charges vs. Tenure')
```
